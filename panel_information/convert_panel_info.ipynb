```python
panel_info_path = "4cast/4cast_amplicon_info.tsv"
```

```python
def convert_panel_info(file_path, output_path):
    panel_info = pd.read_csv(file_path, sep='\t')
    # rename columns 
    panel_info.rename(columns={'ampInsert_start':'insert_start', 'ampInsert_end':'insert_end'}, inplace=True)
    # extract chrom to column
    panel_info['chrom'] = [x[0] for x in panel_info.amplicon.str.split('-')]
    # 0-base coordinates 
    panel_info['insert_start'] = panel_info.insert_start-1
    # make target_id col 
    panel_info['target_id'] = panel_info.chrom + '-' + panel_info.insert_start.astype(str).str.zfill(width=7) + '-' + panel_info.insert_end.astype(str).str.zfill(width=7)
    # conversion_table = panel_info[['target_id','amplicon']]
    # conversion_table.rename(columns={'target_id':'new_name','amplicon':'old_name'}, inplace=True)
    # conversion_table.to_csv(output_path, sep='\t', index=False)
    # drop columns
    panel_info.drop(columns=['amplicon','amplicon_start','amplicon_end','amplicon_length','ampInsert_length'], inplace=True)
    # enforce ordering
    panel_info = panel_info[['chrom','insert_start','insert_end','target_id','fwd_primer','rev_primer']]
    panel_info.to_csv(output_path, sep='\t', index=False)
```
